# Import the required packages in Python

```python
import numpy as np
import pandas as pd
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from collections import Counter
import matplotlib.pyplot as plt
from matplotlib import rcParams
%matplotlib inline
```

# Load the Iris Dataset

```python
data = pd.read_csv('iris.csv')
data.head()
```

# Using counter, checking the count of each sample (Setosa, Versicolor, Virginica)

```python
Counter(data.target)
```

# Defining X and Y

```python
X = data.iloc[:,:-1]
y = data.target
```

# Performing a Train Test Split and checking the accuracy of the model

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, random_state=10)
```

```python
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
y_predict = model.predict(X_test)
accuracy_score(y_test, y_predict)
```

# Using a Crosstab to see how many are correctly and wrongly predicted

```python
pd.crosstab(y_test, y_predict) # Setosa, versicolor, virginica
```

# Visualizing Graphically

```python
rcParams['figure.figsize'] = 18,12
dtc = tree.DecisionTreeClassifier(random_state=10)
dtc = dtc.fit(X,y)
tree.plot_tree(dtc, class_names=['Setosa', 'Versicolor', 'Virginica'], filled=True, rounded=True, feature_names=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])
plt.show()
```
